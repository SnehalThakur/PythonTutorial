```python
firstNameList = []
lastNameList = []
for name in titanic_df.Name:
    if(name.find(",")!=-1):
        names = name.split(",")
        firstNameList.append(names[1])
        lastNameList.append(names[0])
    else:
        firstNameList.append(name)
        lastNameList.append(name)
```

```python
len(firstNameList)
```

```python
len(lastNameList)
```

```python
# titanic_df["First Name"]=firstNameList
titanic_df.insert(loc = 2,
          column = 'First Name',
          value = firstNameList)
```

```python
# titanic_df["Last Name"]= lastNameList
titanic_df.insert(loc = 3,
          column = 'Last Name',
          value = lastNameList)
```
